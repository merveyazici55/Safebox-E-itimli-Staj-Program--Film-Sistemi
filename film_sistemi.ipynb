```python
class FilmSistemi:
    def __init__(self, dosya_adi):
        self.dosya_adi = dosya_adi

    def film_ekle(self):
        film_adi = input("Film Adı: ")
        film_yili = input("Film Yılı: ")
        film_turu = input("Film Türü: ")
        film_yonetmen = input("Film Yönetmeni: ")

        with open(self.dosya_adi, "a") as dosya:
            dosya.write(f"{film_adi}, {film_yili}, {film_turu}, {film_yonetmen}\n")

        print("Film başarıyla eklendi.")

    def film_ara(self):
        film_adi = input("Aranacak Film Adı: ")

        with open(self.dosya_adi, "r") as dosya:
            for satir in dosya:
                film = satir.strip().split(",")
                if film[0] == film_adi:
                    print("Film Bulundu: ")
                    print("Film Adı:", film[0])
                    print("Film Yılı:", film[1])
                    print("Film Türü:", film[2])
                    print("Film Yönetmeni:", film[3])
                    return 
        
        print("Film bulunamadı.")

    def film_sil(self):
        film_adi = input("Silinecek Film Adi: ")
        bulundu = False

        with open(self.dosya_adi, "r") as dosya:
            filmler = dosya.readlines()

        with open(self.dosya_adi, "w") as dosya:
            for satir in filmler:
                film = satir.strip().split(",")
                if film[0] == film_adi:
                    bulundu = True
                else:
                    dosya.write(satir)

        if bulundu :
            print("Film başarıyla silindi.")
        else:
            print("Film bulunamadı.")


sistem = FilmSistemi("filmler.txt")

while True:
    print("[1] Film Ekle")
    print("[2] Film Ara")
    print("[3] Film Sil")
    print("[4] Çıkış")

    secim = input("Seçimizini yapın(1-4):")

    if secim == "1":
        sistem.film_ekle()
    elif secim == "2":
        sistem.film_ara()
    elif secim == "3":
        sistem.film_sil()
    elif secim == "4":
        break
    else:
        print("Geçersiz seçim yaptınız. Tekrar deneyin.")
```
